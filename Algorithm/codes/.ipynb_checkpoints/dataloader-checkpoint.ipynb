```python
from sklearn.model_selection import train_test_split
```

```python
anor_test = '../HTTP_2010/anomalousTrafficTest.txt'
nor_test = '../HTTP_2010/normalTrafficTest.txt'
nor_train = '../HTTP_2010/normalTrafficTraining.txt'

with open(anor_test, 'r') as f:
    file = f.read()
cont = file.split('\n\n')
anor_test_data = [d.strip('\n') for d in cont if d.startswith(("GET", "POST", "PUT"))]

with open(nor_test, 'r') as f:
    file = f.read()
cont = file.split('\n\n')
nor_test_data = [d.strip('\n') for d in cont if d.startswith(("GET", "POST", "PUT"))]

with open(nor_train, 'r') as f:
    file = f.read()
cont = file.split('\n\n')
nor_train_data = [d.strip('\n') for d in cont if d.startswith(("GET", "POST", "PUT"))]

data = anor_test_data + nor_test_data + nor_train_data
label = [1] * len(anor_test_data) + [0] * (len(nor_test_data)+len(nor_train_data))
assert len(data) == len(label)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, label, test_size=0.2, shuffle=True)
print("len(x_train): %d" % len(x_train))
print("len(y_train): %d" % len(y_train))
print(x_train[:2])
print(y_train[:2])

print("len(x_test): %d" % len(x_test))
print("len(y_test): %d" % len(y_test))
print(x_test[:2])
print(y_test[:2])
```
